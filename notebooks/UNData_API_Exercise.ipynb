1. Use the API to get all available data for the _GDP per capita, PPP (constant 2017 international $)_ indicator. Hint: this indicator has code "NY.GDP.PCAP.PP.KD". Adjust the query parameters so that you can retrieve all available rows. Convert the results to a DataFrame.

> ### Convert API to Dataframe
>
> Below we have to different API's. We will use keywords from the API query instructions to modify our API call. This way we get only the nesecarry information for analysis. This keeps the data consistent and easy for other people to work with. YOu will notice ive don a few different things.
>
>* Make **Specific** API calls
>* Filter and sort **JSON** formatted data
>* **Clean and Merge** data for **insightful analysis.**


```python
endpoint = 'https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD?format=json&per_page=20000'

response = requests.get(endpoint)
```

```python
response
```

```python
response.text[:500]
```

```python
res = response.json()
res[:1]
```

```python
# Retrieve keys from dictionary
res[1][0].keys()
```

```python
gdp_df = pd.json_normalize(res[1])
gdp_df.head(2)
```

2. Now, use the API to get all available data for _Life expectancy at birth, total (years)_. This indicator has code "SP.DYN.LE00.IN". Again, convert the results to a DataFrame.

>  This is the same process as above. We are making a **second API call.** This data will contain **life expectancy values.**

```python
endpoint2 = 'https://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN?format=json&per_page=20000'

response2 = requests.get(endpoint2)
```

```python
response2
```

```python
response2.text[:500]
```

```python
res2 = response2.json()
res2[:1]
```

```python
# Retrieve keys from dictionary
res2[1][0].keys()
```

```python
le_df = pd.json_normalize(res2[1])
le_df.head(2)
```

3. Merge the two results DataFrames together. You may want to rename or drop columns prior to merging.

> ### Merge and Clean Data
>
> In these last questions we have sucesfully pulled data from the UN Data API and condensed json files into dataframes. With this information we can gather much valuable information, such as interest rates, life expectancy and GDP. Next, we are going to merge all the information we have gathered into one data frame. Below i have done a few things
>
>* Re-name ambiguous columns in the dataframe with **identifiable names.**
>* We have **dropped empty columns** containing excess NaN values.
>* Merge **all information** for **one final, clean dataframe** for exploration.

```python
gdp_df = gdp_df.rename(columns={'value': 'gdp_per_capita_ppp'})
```

```python
le_df = le_df.rename(columns={'value': 'life_expectancy'})
```

```python
merged_gdp_le = pd.merge(gdp_df, le_df, on=['countryiso3code', 'date'], how='inner')
```

```python
clean_gdp_le_df = merged_gdp_le[['countryiso3code', 'date', 'gdp_per_capita_ppp', 'life_expectancy', 'country.value_x']].rename(columns={'countryiso3code': 'country_code', 'country.value_x': 'country'})
```

```python
clean_gdp_le_df.head()
```
